### Задача 4. Кожну клітинку квадратної таблиці 2×2 можна пофарбувати у чорний чи білий колір. Скільки існує різних кольорів цієї таблиці?

```python
# кількість клітин
number_of_cells = 4
# кількість кольорів
number_of_colors = 2
```

N = k^n

```python
result = pow(number_of_colors, number_of_cells ) 
```

```python
print(f"{result} різних розфарбувань")
```

### Задача 4. Скільки способами можна побудувати замкнуту ламану, вершинами якої є вершини правильного шестикутника (ламана може бути самоперетинається)? 


```python
# кількість вершин шестикутника
n_vertices = 6
```

```python
N = (n_vertices - 1) // 1
```

```python
result = factorial(n_vertices - 1) // 2
```

```python
print(f"{result} способів побудувати замкнуту ламану")
```
