```python
# Зчитуємо CSV
file_path = "./student_exam_scores.csv"
df = pd.read_csv(file_path)

x = df["sleep_hours"]
y = df["exam_score"]

# Обчислюємо середні значення
x_mean = np.mean(x)
y_mean = np.mean(y)

# -------- 1. Кореляція (r) --------
# формула: r = Σ((x - x̄)(y - ȳ)) / sqrt(Σ(x - x̄)^2 * Σ(y - ȳ)^2)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))
r = numerator / denominator

print(f"Коефіцієнт кореляції (r): {r:.4f}")

# -------- 2. Лінійна регресія (y = b0 + b1*x) --------
# формули:
# b1 = Σ((x - x̄)(y - ȳ)) / Σ((x - x̄)^2)
# b0 = ȳ - b1*x̄
b1 = numerator / np.sum((x - x_mean)**2)
b0 = y_mean - b1 * x_mean

print(f"Рівняння регресії: y = {b0:.2f} + {b1:.2f}x")

# -------- 3. Коефіцієнт детермінації --------
# R² = r²
R2 = r**2
print(f"Коефіцієнт детермінації (R²): {R2:.4f}")

# -------- 4. Побудова графіка --------
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Дані')
plt.plot(x, b0 + b1*x, color='red', label='Лінія регресії')
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Score")
plt.title("Регресія: Sleep Hours vs Exam Score (все руками)")
plt.legend()
plt.grid(True)
plt.show()
```
